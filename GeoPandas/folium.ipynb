```python
import pandas as pd
import folium
import geopandas as gpd
```

```python
location = "https://data.smartdublin.ie/dataset/33ec9fe2-4957-4e9a-ab55-c5e917c7a9ab/resource/2dec86ed-76ed-47a3-ae28-646db5c5b965/download/dublin.csv"
bike_station_locations = pd.read_csv(location)
```

```python
bike_station_locations
```

<b>Me quedo con las posiciones</b>

```python
bike_station_locations = bike_station_locations[["Latitude", "Longitude", "Name"]]
```

```python
bike_station_locations
```

```python
map = folium.Map(location=[bike_station_locations.Latitude.mean(), bike_station_locations.Longitude.mean()], zoom_start=14, control_scale=True)
```

```python
map
```

<b>Cargo los puntos al mapa</b>

```python
for index, location_info in bike_station_locations.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Name"]).add_to(map)
```

```python
map
```

<b>Segundo Ejemplo - Datos de olas en diferentes puertos</b>

```python
import json

import requests

url = (
    "https://raw.githubusercontent.com/python-visualization/folium/main/examples/data"
)
vis1 = json.loads(requests.get(f"{url}/vis1.json").text)
vis2 = json.loads(requests.get(f"{url}/vis2.json").text)
vis3 = json.loads(requests.get(f"{url}/vis3.json").text)
```

```python
vis3
```

```python
m = folium.Map(location=[46.3014, -123.7390], zoom_start=7, tiles="Stamen Terrain")

folium.Marker(
    location=[47.3489, -124.708],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(vis1, width=450, height=250)
    ),
).add_to(m)

folium.Marker(
    location=[44.639, -124.5339],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(vis2, width=450, height=250)
    ),
).add_to(m)

folium.Marker(
    location=[46.216, -124.1280],
    popup=folium.Popup(max_width=450).add_child(
        folium.Vega(vis3, width=450, height=250)
    ),
).add_to(m)


m
```

```python
archivo_shp = 'PaisLocalidades.shp'

gdf = gpd.read_file(archivo_shp)
gdf.to_crs(4326)

geojson = gdf.to_json()
centroide = gdf.geometry.centroid.iloc[0]

mapa = folium.Map(location=[centroide.y, centroide.x], zoom_start=8)

folium.GeoJson(geojson).add_to(mapa)

mapa
mapa.save("mapa_de_urugay.html")
```

```python
import folium
import pandas as pd

eco_footprints = pd.read_csv("datos.csv")
political_countries_url = (
    "http://geojson.xyz/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson"
)

m = folium.Map(location=(30, 10), zoom_start=3, tiles="cartodb positron")
folium.Choropleth(
    geo_data=political_countries_url,
    data=eco_footprints,
    columns=["Country/region", "Ecological footprint"],
    key_on="feature.properties.name",
).add_to(m)

m.save("footprint.html")
```

```python
import folium
import pandas as pd

eco_footprints = pd.read_csv("datos.csv")
political_countries_url = (
    "http://geojson.xyz/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson"
)

m = folium.Map(location=(30, 10), zoom_start=3, tiles="cartodb positron")
folium.Choropleth(
    geo_data=political_countries_url,
    data=eco_footprints,
    columns=("Country/region", "Ecological footprint"),
    key_on="feature.properties.name",
    fill_color="RdYlGn_r",
    fill_opacity=0.8,
    line_opacity=0.3,
    nan_fill_color="white",
    legend_name="Ecological footprint per capita",
).add_to(m)

m
```
