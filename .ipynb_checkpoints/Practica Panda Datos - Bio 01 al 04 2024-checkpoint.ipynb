```python
# Extraer los nombres de lugares únicos como una lista
lista_lugares_unicos = df['Nombre Cliente'].unique().tolist()
print(lista_lugares_unicos)
```

```python
import csv

#Lugares lista
nombres_lugares = lista_lugares_unicos

# Nombre del archivo CSV de salida
nombre_archivo = "lugares_uy.csv"

# Escribir los nombres de lugares en el archivo CSV
with open(nombre_archivo, 'w', newline='', encoding='utf-8') as archivo:
    escritor_csv = csv.writer(archivo)
    for lugar in nombres_lugares:
        escritor_csv.writerow([lugar])

print("Archivo CSV generado correctamente:", nombre_archivo)
```

```python
df_lugares = pd.read_csv("lugares_uy.csv", sep=';', encoding='latin-1', on_bad_lines='skip')
df_lugares
```

```python
# Realizar el join entre los DataFrames en la columna 'Nombre_cliente'
df_merge = pd.merge(df, df_lugares, on='Nombre Cliente', how='inner')
df_merge
```

```python
df_merge.columns
```

```python
#Conteo de NDs por deparamento
conteo_por_deparamento = df_merge.groupby('Departamento')['Nombre Cliente'].count()
conteo_por_deparamento
```

```python
# Crear un DataFrame a partir de la serie
df_mostrar_mapa = pd.DataFrame({'Departamento': conteo_por_deparamento.index ,'Cantidad': conteo_por_deparamento.values })
def obtener_latitud_longitud(departamento): 
    pais = "Uruguay"
    limites = ((-34.953, -57.76), (-30.107, -53.68))  # Coordenadas del límite de Uruguay
    try:
        # Realizar la búsqueda de empresas dentro de Uruguay
        address = geo.geocode(departamento+", Uruguay",timeout=0.5)
        return address.latitude,address.longitude
    except:
        return None , None

# Aplicar la función para obtener latitud y longitud y crear nuevas columnas
df_mostrar_mapa['Latitud'], df_mostrar_mapa['Longitud'] = zip(*df_mostrar_mapa['Departamento'].apply(obtener_latitud_longitud))

df_mostrar_mapa
df_mostrar_mapa.columns
```

```python
df_mostrar_mapa
```

```python
import pydeck as pdk

layer = pdk.Layer(
    'ScatterplotLayer',  # `type` positional argument is here
    df_mostrar_mapa,
    get_position=['Longitud', 'Latitud'],
    auto_highlight=True,
    get_radius=10000,          # Radius is given in meters
    get_fill_color=[180, 0, 200, 140],  # Set an RGBA value for fill
    pickable=True)

# Set the viewport location
view_state = pdk.ViewState(
    longitude=df_mostrar_mapa['Longitud'].mean(),
    latitude=df_mostrar_mapa['Latitud'].mean(),
    zoom=6,
    min_zoom=5,
    max_zoom=15,
    pitch=40.5,
    bearing=-27.36)


# Combined all of it and render a viewport
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.to_html('nd-departamentos-ejemplo.html')
```
