```python
conteo_horas_por_cliente = df.groupby('Nombre Cliente')['Tiempo efectivo'].sum()
conteo_horas_por_cliente = conteo_horas_por_cliente.sort_values(ascending = False)
conteo_horas_por_cliente
```

```python
# Graficar el resultado para el TOP 10 de Mayor tiempo
conteo = conteo_horas_por_cliente[:5]
plt.figure(figsize=(10, 6))
plt.bar(conteo.index, conteo.values, color='limegreen')
plt.title('Top 5 Cantidad de horas dedicas por cliente', fontsize=16)
plt.xlabel('Cliente', fontsize=14)
plt.ylabel('Horas', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
conteo_horas_por_equipo = df.groupby('Nombre Equipo')['Tiempo efectivo'].sum()
conteo_horas_por_equipo = conteo_horas_por_equipo.sort_values(ascending = False)
conteo_horas_por_equipo
```

```python
# Graficar el resultado para el TOP 5
# Seleccionar un colormap
colormap = plt.cm.tab10
# N칰mero de colores que deseas generar
numero_colores = 5
# Generar una lista de colores muestreando el colormap
lista_colores = [colormap(i) for i in range(numero_colores)]
conteo = conteo_horas_por_equipo[:5]
plt.figure(figsize=(10, 6))
plt.bar(conteo.index, conteo.values, color=lista_colores)
plt.title('Top 5 Cantidad de horas dedicas por Equipos', fontsize=16)
plt.xlabel('Equipo', fontsize=14)
plt.ylabel('Horas', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
