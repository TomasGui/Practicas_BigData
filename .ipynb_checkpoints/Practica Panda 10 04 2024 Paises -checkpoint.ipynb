```python
df1 = df.sort_values(by='child_mort', ascending = False)
top_10 = df1.head(10)

# Crear la grÃ¡fica
plt.figure(figsize=(10, 6))
plt.bar(top_10['country'], top_10['child_mort'], color='skyblue')
plt.title('Top 10 PaÃ­ses con Mayor Mortalidad Infantil', fontsize=16)
plt.xlabel('PaÃ­s', fontsize=14)
plt.ylabel('Cantidad de Muertes', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```

```python
df1 = df.sort_values(by='life_expec', ascending = False)
top_10 = df1.head(10)

# Crear la grÃ¡fica
plt.figure(figsize=(10, 6))
plt.bar(top_10['country'], top_10['life_expec'], color='skyblue')
plt.title('Top 10 PaÃ­ses con Expectativa de vida', fontsize=16)
plt.xlabel('PaÃ­s', fontsize=14)
plt.ylabel('AÃ±os de vida', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```

```python
df2 = df.sort_values(by='inflation', ascending = True)
top_10 = df2.head(10)

# Crear la grÃ¡fica
plt.figure(figsize=(10, 6))
plt.bar(top_10['country'], top_10['inflation'], color='skyblue')
plt.title('Top 10 PaÃ­ses con menor inflacion', fontsize=16)
plt.xlabel('PaÃ­s', fontsize=14)
plt.ylabel('Inflacion', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```
