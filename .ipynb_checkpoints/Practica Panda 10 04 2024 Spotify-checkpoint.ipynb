```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Popular_Spotify_Songs.csv", encoding='latin-1')
```

```python
df
```

```python
print(df.dtypes)
```

```python
# Convertir la columna 'FechaEntera' a tipo datetime
df['released_year'] = pd.to_datetime(df['released_year'], format='%Y')
df['released_month'] = pd.to_datetime(df['released_month'], format='%m')
df['released_day'] = pd.to_datetime(df['released_day'], format='%d')
print(df.dtypes)
```

```python
cantidad_por_año = df.groupby('released_year')['artist(s)_name'].count()
cantidad_por_año 
```

```python
# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(cantidad_por_año.index, cantidad_por_año.values, color='skyblue')
plt.title('Cantidad de musicas producidas por Año', fontsize=16)
plt.xlabel('Años', fontsize=14)
plt.ylabel('Cantidad', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```

```python
cantidad_por_mode = df.groupby('mode')['artist(s)_name'].count()
cantidad_por_mode
```

```python
# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(cantidad_por_mode.index, cantidad_por_mode.values, color=['red', 'blue'])
plt.title('Cantidad de musicas por Mode', fontsize=16)
plt.xlabel('Mode', fontsize=14)
plt.ylabel('Cantidad', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
```

```python
cantidad_por_año_mes = df.groupby(['released_year' , 'released_month'])['artist(s)_name'].count()
cantidad_por_año_mes
```

```python
# Histograma

plt.figure(figsize=(10, 6))
plt.hist(cantidad_por_año_mes.values, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Títulos por Año', fontsize=16)
plt.xlabel('Año', fontsize=14)
plt.ylabel('Cantidad de Títulos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
